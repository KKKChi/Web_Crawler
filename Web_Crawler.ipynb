```python
def getinfo(URL):
    
    driver.get(URL)
    
    Parcel1 = driver.find_element_by_xpath('//*[@id="parcel-views"]/div[1]/div[2]/div/div[1]/div[1]/div[2]')
    Parcel = Parcel1.text
    
    Info1 = driver.find_elements_by_class_name('col-sm-8')
    list1 = list()
    for i in Info1:
        list1.append(i.text)
    Owner = list1[0]
    fulladdress = list1[1]
    Owneradd_line1 = re.findall(r'(.*)\n.*',fulladdress)
    Owneradd_line2 = re.findall(r'.*\n(.*)',fulladdress)
    Owneradd_line1_fin = ''.join(Owneradd_line1)
    Owneradd_line2_fin = ''.join(Owneradd_line2)
    
    Psy_address1 = driver.find_element_by_xpath('//*[@id="parcel-views"]/div[1]/div[2]/div/div[1]/div[6]/div[2]')
    Psy_address = Psy_address1.text
    Acre1 = driver.find_element_by_xpath('//*[@id="parcel-views"]/div[1]/div[2]/div/div[2]/div[1]/div[2]')
    Acre = Acre1.text
    Legal_des1 = driver.find_element_by_xpath('//*[@id="parcel-views"]/div[5]/table/tbody/tr/td')
    Leg_des = Legal_des1.text
    Class1 = driver.find_element_by_xpath('//*[@id="parcel-views"]/div[1]/div[2]/div/div[1]/div[3]/div[2]')
    Zoning = Class1.text
    tax_del_2019_1 = driver.find_element_by_xpath('//*[@id="parcel-views"]/div[4]/table/tbody[1]/tr[1]/td[4]')
    tax_del_2019 = tax_del_2019_1.text
    tax_del_2018_1 = driver.find_element_by_xpath('//*[@id="parcel-views"]/div[4]/table/tbody[1]/tr[2]/td[4]')
    tax_del_2018 = tax_del_2018_1.text
    Property_class1 = driver.find_element_by_xpath('//*[@id="parcel-views"]/div[10]/table/tbody/tr/td[1]')
    Property_class = Property_class1.text
    Land_Value1 = driver.find_element_by_xpath('//*[@id="parcel-views"]/div[12]/table/tbody/tr[1]/td[2]')
    Land_Value = Land_Value1.text
    Building_Value1 = driver.find_element_by_xpath('//*[@id="parcel-views"]/div[12]/table/tbody/tr[1]/td[3]')
    Building_Value = Building_Value1.text
    Total_Value1 = driver.find_element_by_xpath('//*[@id="parcel-views"]/div[12]/table/tbody/tr[1]/td[4]')
    Total_Value = Total_Value1.text
    
    Infolist = list()
    Infolist.append(Parcel)
    Infolist.append(Owner)
    Infolist.append(Owneradd_line1_fin)
    Infolist.append(Owneradd_line2_fin)
    Infolist.append(Psy_address)
    Infolist.append(Leg_des)
    Infolist.append(Zoning)
    Infolist.append(tax_del_2019)
    Infolist.append(tax_del_2018)
    Infolist.append(Property_class)
    Infolist.append(Land_Value)
    Infolist.append(Building_Value)
    Infolist.append(Total_Value)
    
    return Infolist
```
