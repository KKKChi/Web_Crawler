```python
def getinfo(URL):
    
    driver.get(URL)
    
    Parcel1 = driver.find_element_by_xpath('//*[@id="dnn_ctr416_View_tdPropertyID"]')
    Parcel = Parcel1.text
    PropertyAddress1 = driver.find_element_by_xpath('//*[@id="dnn_ctr416_View_tdPropertyAddress"]')
    PropertyAddress = PropertyAddress1.text
    Owner1 = driver.find_element_by_xpath('//*[@id="dnn_ctr416_View_divOwnersLabel"]')
    Owner = Owner1.text
    
    MailingAddressfull1 = driver.find_element_by_xpath('//*[@id="dnn_ctr416_View_tdOIMailingAddress"]')
    MailingAddressfull = MailingAddressfull1.text
    MailingAddressLine1_1 = re.findall(r'(^\S.*)\s\w\S*,',MailingAddressfull)
    MailingAddressLine1 = ''.join(MailingAddressLine1_1)
    MailingAddressCity1 = re.findall(r'\s(\w\S*),',MailingAddressfull)
    MailingAddressCity = ''.join(MailingAddressCity1)
    MailingAddressState1 = re.findall(r',\s(\w{2})',MailingAddressfull)
    MailingAddressState = ''.join(MailingAddressState1)
    MailingAddressZip1 = re.findall(r',\s\w{2}\s(\d.*$)',MailingAddressfull)
    MailingAddressZip = ''.join(MailingAddressZip1)
    
    Zoning1 = driver.find_element_by_xpath('//*[@id="dnn_ctr416_View_tdGIPropertyType"]')
    Zoning = Zoning1.text
    Leg_Des1 = driver.find_element_by_xpath('//*[@id="dnn_ctr416_View_tdGILegalDescription"]')
    Leg_Des = Leg_Des1.text
    Neighborhood1 = driver.find_element_by_xpath('//*[@id="dnn_ctr416_View_tdGINeighborhood"]')
    Neighborhood = Neighborhood1.text
    Property_status1 = driver.find_element_by_xpath('//*[@id="dnn_ctr416_View_tdGIPropertyStatus"]')
    Status = Property_status1.text
    BuildingValue1 = driver.find_element_by_xpath('//*[@id="dnn_ctr416_View_tdVITotalImprovementMV"]')
    BuildingValue = BuildingValue1.text
    LandValue1 = driver.find_element_by_xpath('//*[@id="dnn_ctr416_View_tdVITotalLandMV"]')
    LandValue = LandValue1.text
    TotalValue1 = driver.find_element_by_xpath('//*[@id="dnn_ctr416_View_tdVITotalAssessedValueRP"]')
    TotalValue = TotalValue1.text
    DeedDate1 = driver.find_element_by_xpath('//*[@id="dnn_ctr416_View_tblSalesHistoryData"]/tbody/tr[2]/td[1]')
    DeedDate = DeedDate1.text
    
    Infolist = list()
    Infolist.append(Parcel)
    Infolist.append(PropertyAddress)
    Infolist.append(Owner)
    Infolist.append(MailingAddressLine1)
    Infolist.append(MailingAddressCity)
    Infolist.append(MailingAddressState)
    Infolist.append(MailingAddressZip)
    Infolist.append(Zoning)
    Infolist.append(Leg_Des)
    Infolist.append(Neighborhood)
    Infolist.append(Status)
    Infolist.append(BuildingValue)
    Infolist.append(LandValue)
    Infolist.append(TotalValue)
    Infolist.append(DeedDate)
    
    return Infolist
```
