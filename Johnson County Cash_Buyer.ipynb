```python
from selenium import webdriver
```

```python
drive = webdriver.Chrome('/Users/key.gao/Downloads/chromedriver')
```

```python
def getinfo(URL):
    
    drive = webdriver.Chrome('/Users/key.gao/Downloads/chromedriver')

    drive.get(URL)
    
    n3 = 'col-3'
    n4 = 'col-4'
    n5 = 'col-5'
    n6 = 'col-6'
    n7 = 'col-7'
    n8 = 'col-8'
    n9 = 'col-9'

    Grantor = drive.find_element_by_class_name(n3).text
    Grantee = drive.find_element_by_class_name(n4).text
    Doc_Type = drive.find_element_by_class_name(n5).text
    Doc_Date = drive.find_element_by_class_name(n6).text
    Doc_Number = drive.find_element_by_class_name(n7).text
    Book_Volumn_Page = drive.find_element_by_class_name(n8).text
    Legal_Desc = drive.find_element_by_class_name(n9).text
    
    infolist = list()
    infolist.append(Grantor)
    infolist.append(Grantee)
    infolist.append(Doc_Type)
    infolist.append(Doc_Date)
    infolist.append(Doc_Number)
    infolist.append(Book_Volumn_Page)
    infolist.append(Legal_Desc)
    drive.quit()
    return infolist
```

```python
getinfo("https://johnson.tx.publicsearch.us/results?department=RP&recordedDateRange=17710617%2C20200430&searchOcrText=false&searchType=quickSearch&searchValue=2020-9998")
```

```python
PreURL = "https://johnson.tx.publicsearch.us/results?department=RP&recordedDateRange=17710617%2C20200430&searchOcrText=false&searchType=quickSearch&searchValue="
```

```python
URLlist = list()
for i in range(0,10000):
    URL_R1 = "-"+str(i).zfill(4)
    for o in range(2016,2021):
        URL_R = str(o)+URL_R1
        URL=PreURL+URL_R
        URLlist.append(URL)
```

```python
URLlist= URLlist[::-1]
```

```python
from selenium.common.exceptions import NoSuchElementException
```

```python
result = list()
for URL in URLlist[44534:50000]:
    try:
        Res = getinfo(URL)
    except NoSuchElementException as NSEE:
        drive.quit()
        print(URL+" NSEE")
        pass
    else:
        print(URL+" finished")
        result.append(Res)
```
