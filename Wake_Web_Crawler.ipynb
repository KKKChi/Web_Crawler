```python
# Comb1 getinfo fuction
def getinfo(web_address):
    html = urlopen(web_address)
    bs = BeautifulSoup(html.read(), 'html.parser')
    #APN
    APNraw = bs.find('title').get_text()
    APN = re.findall(r'0\d{6}',APNraw)
    APN = ''.join(APN)
    
    highlightInformation = bs.find_all('font',{'size':'2','face':'Arial'})
    highlightInformation
    Infolist = list()
    for i in highlightInformation:
        Infolist.append(i.get_text())
    Owner = Infolist[16]
    OwnerAddressLine1 = Infolist[20]
    OwnerAddressLine2 = Infolist[21]
    PropertyAddressLine1 = Infolist[24]
    
    PropertyAddressLine2raw = Infolist[25]
    PropertyAddressLine2_1 = re.findall(r'(\S*)\xa0NC\xa0',PropertyAddressLine2raw)
    PropertyAddressLine2_2 = re.findall(r'\xa0NC\xa0(\d{5}-\d{4})',PropertyAddressLine2raw)
    PropertyAddressLine2_1.append('NC')
    PropertyAddressLine2_all = PropertyAddressLine2_1 + PropertyAddressLine2_2
    PropertyAddressLine2 = " ".join(PropertyAddressLine2_all)
    
    Legal_Description= Infolist[12]
    Land_Class = Infolist[41]
    Zoning = Infolist[47]
    Township = Infolist[39]
    Acreage = Infolist[53]
    Deed_date = Infolist[61]
    Pkg_Price = Infolist[69]
    Land_Price = Infolist[73]
    Land_Assessed_Value = Infolist[88]
    Building_Assessed_Value = Infolist[90]
    Total_Assessed_Value = Infolist[114]
    
    inforow = list()
    inforow.append(APN)
    inforow.append(Owner)
    inforow.append(OwnerAddressLine1)
    inforow.append(OwnerAddressLine2)
    inforow.append(PropertyAddressLine1)
    inforow.append(PropertyAddressLine2)
    inforow.append(Legal_Description)
    inforow.append(Land_Class)
    inforow.append(Zoning)
    inforow.append(Township)
    inforow.append(Acreage)
    inforow.append(Deed_date)
    inforow.append(Pkg_Price)
    inforow.append(Land_Price)
    inforow.append(Land_Assessed_Value)
    inforow.append(Building_Assessed_Value)
    inforow.append(Total_Assessed_Value)
    
    return inforow
```
